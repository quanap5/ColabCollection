```python
import pandas as pd
import numpy as np

# Create a dataframe with midterm grades for 10 students
data = {'Student': [f'Student_{i+1}' for i in range(10)],
        'Midterm_Grade': np.random.randint(60, 101, size=10)}

df_grades_new = pd.DataFrame(data)

display(df_grades_new)
```

```python
# @title Midterm_Grade

from matplotlib import pyplot as plt
df_grades_new['Midterm_Grade'].plot(kind='line', figsize=(8, 4), title='Midterm_Grade')
plt.gca().spines[['top', 'right']].set_visible(False)
```
